```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x, y = np.loadtxt("profiles.csv", delimiter=',', unpack=True)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#normalise data so that we can ask for the 50% line later
ynorm = y / y.max() * 100

def smooth(y, box_pts):
    #create a convolution of size box_pts
    box = np.ones(box_pts)/box_pts
    print np.ones(box_pts),box
    #smooth data by convolving with box created from box_pts value
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

ynorm = smooth(ynorm, 50)
plt.scatter(x,ynorm)
```

```python
from scipy.interpolate import UnivariateSpline
spline = UnivariateSpline(x,ynorm-50,s=0)
fwhm = abs(spline.roots()[1]-spline.roots()[0])
print fwhm.round(3)
```
